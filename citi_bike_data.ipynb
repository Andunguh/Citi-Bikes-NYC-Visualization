```python
import pandas as pd
from functools import reduce
```

```python
# compile the list of dataframes you want to merge
df1 = pd.read_table("Data/January2020.csv", sep=',')
df2 = pd.read_table("Data/February2020.csv", sep=',')
df3 = pd.read_table("Data/March2020.csv", sep=',')
df4 = pd.read_table("Data/April2020.csv", sep=',')
df5 = pd.read_table("Data/May2020.csv", sep=',')
df6 = pd.read_table("Data/June2020.csv", sep=',')
df7 = pd.read_table("Data/July2020.csv", sep=',')
df8 = pd.read_table("Data/August2020.csv", sep=',')
df9 = pd.read_table("Data/September2020.csv", sep=',')
df10 = pd.read_table("Data/October2020.csv", sep=',')
df11 = pd.read_table("Data/November2020.csv", sep=',')
df12 = pd.read_table("Data/December2020.csv", sep=',')
```

```python
# compile the list of dataframes you want to merge
data_frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,
                                            how='outer'), data_frames)
```
